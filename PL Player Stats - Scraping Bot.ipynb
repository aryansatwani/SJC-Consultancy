# PL player data

Import the Required Modules

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
import numpy as np
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)
```

```python
#Change website based for other leagues
driver.get("https://fbref.com/en/comps/12/stats/La-Liga-Stats")
driver.maximize_window()
time.sleep(10)
```

```python
elements = driver.find_elements(By.XPATH,"//table[@id='stats_standard']/tbody/tr/td") # 36 per stat
```

```python
dat = []
for i in elements:
    dat.append(i.text)
```

```python
i=0
dat[i:i+36]
```

```python
#main list creation
player_data = [dat[i:i+36] for i in range(0,len(dat), 36)]
```

```python
player_data
```

```python
#per 90 minutes stats written as {stat}/90
titles = "Rk Player Nation Pos Squad Age Born MP Starts Min 90s Gls Ast G+A G-PK PK PKatt CrdY CrdR xG npxG xAG npxG+xAG PrgC PrgP PrgR Gls/90 Ast/90 G+A/90 G-PK/90 G+A-PK/90 xG/90 xAG/90 xG+xAG/90 npxG/90 npxG+xAG/90 Matches"
column_titles = titles.split(' ')
column_titles = column_titles[1:]
```

```python
len(column_titles)
```

```python
df=pd.DataFrame(player_data,columns=column_titles)
```

```python
df.head()
```

Cleaning the dataset into correct format

```python
df['Nation'].unique()
```

```python
# Nation was written twice in small and Caps hence removed the small.
df['Nation'] = df['Nation'].apply(lambda x:x.split(' ')[1] if len(str(x))>0 else x)
# Minutes had commas so couldnt convert to int
df['Min']=df['Min'].apply(lambda x:x.replace(',',''))
# unwanted column
df.drop('Matches',axis=1,inplace=True)
# certain stats were empty strings replaces with nan
df.replace('',np.nan,inplace=True)
#only taking main position assuming its the first one.
df['Pos'] = df['Pos'].apply(lambda x:x.split(',')[0])
#age only years
df['Age'] = df['Age'].apply(lambda x:x.split('-')[0] if str(x)!='nan' else x)
```

```python
df.drop(df[df['Player']=='Yehor Yarmoliuk'].index,inplace=True)
```

Converting the data into the correct type

```python
try: 
    df['Born'] = df['Born'].astype(int)
    df['MP'] = df['MP'].astype(int)
    df['Starts'] = df['Starts'].astype(int)
    df['Min'] = df['Min'].astype(int)
    df['90s'] = df['90s'].astype(float)
    df['Gls'] = df['Gls'].astype(int)
    df['Ast'] = df['Ast'].astype(int)
    df['G+A'] = df['G+A'].astype(int)
    df['G-PK'] = df['G-PK'].astype(int)
    df['PK'] = df['PK'].astype(int)

    df['PKatt'] = df['PKatt'].astype(int)
    df['CrdY'] = df['CrdY'].astype(int)
    df['CrdR'] = df['CrdR'].astype(int)
    df['xG'] = df['xG'].astype(float)
    df['npxG'] = df['npxG'].astype(float)
    df['xAG'] = df['xAG'].astype(float)
    df['npxG+xAG'] = df['npxG+xAG'].astype(float)
    
    df['PrgC'] = df['PrgC'].astype(int)
    df['PrgP'] = df['PrgP'].astype(int)
    df['PrgR'] = df['PrgR'].astype(int)
    
    df['Gls/90'] = df['Gls/90'].astype(float)
    df['Ast/90'] = df['Ast/90'].astype(float)
    df['G+A/90'] = df['G+A/90'].astype(float)
    df['G-PK/90'] = df['G-PK/90'].astype(float)
    df['G+A-PK/90'] = df['G+A-PK/90'].astype(float)
    df['xG/90'] = df['xG/90'].astype(float)
    df['xAG/90'] = df['xAG/90'].astype(float)
    df['xG+xAG/90'] = df['xG+xAG/90'].astype(float)
    df['npxG/90'] = df['npxG/90'].astype(float)
    df['npxG+xAG/90'] = df['npxG+xAG/90'].astype(float)
except:
    pass
```

```python
df.info()
```

### Saving Dataframe

```python
df.to_csv('LaLiga_player_attacking_data_2023-24.csv',index=False)
```

```python
titles = "Rk Player Nation Pos Squad Age Born MP Starts Min 90s GA GA90 SoTA Saves Save% W D L CS CS% PKatt PKA PKsv PKm PSave% Matches"
column_titles = titles.split(' ')
column_titles = column_titles[1:]
```

```python
# Nation was written twice in small and Caps hence removed the small.
df['Nation'] = df['Nation'].apply(lambda x:x.split(' ')[1] if ' ' in x else x)
# unwanted column
df.drop('Matches',axis=1,inplace=True)
# certain stats were empty strings replaces with nan
df.replace('',np.nan,inplace=True)
#age only years
df['Age'] = df['Age'].apply(lambda x:x.split('-')[0])
df['Min']=df['Min'].apply(lambda x:x.replace(',',''))
```
